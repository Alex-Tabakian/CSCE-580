```python
# full_pipeline_compat_inmemory_fixed.py
# In-memory pipeline (no file I/O). Robust and Colab-friendly.

import sys, subprocess, time, warnings, inspect, os
from pathlib import Path
from typing import List
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch

# Auto-install (lightweight: don't pin exact versions)
_try_install = True
if _try_install:
    required_pkgs = ['transformers', 'datasets', 'evaluate', 'torch', 'scikit-learn', 'pandas', 'matplotlib', 'seaborn']
    for p in required_pkgs:
        try:
            __import__(p)
        except Exception:
            subprocess.check_call([sys.executable, "-m", "pip", "install", p])

from transformers import (
    AutoTokenizer,
    AutoModel,
    AutoModelForSequenceClassification,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer
)
from datasets import Dataset as HFDataset
import evaluate

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split

# Notebook display
try:
    from IPython.display import display
    _HAS_IPY = True
except Exception:
    _HAS_IPY = False

# -------------------------
# Config
# -------------------------
DATA_PATH = "/content/IMDB Dataset.csv"
FAST = False
SEED = 42
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
DISTIL_MODEL = 'distilbert-base-uncased'
EPOCHS = 1 if FAST else 15
BATCH_SIZE = 16 if torch.cuda.is_available() else 8
LR = 2e-5
MAX_LENGTH = 192
GPT2_MODEL = 'gpt2'
SHOW_PLOTS = True

os.environ["WANDB_DISABLED"] = "true"

# -------------------------
# Utilities
# -------------------------
def set_seed(seed=42):
    import random
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
set_seed(SEED)

def _is_colab():
    return 'google.colab' in sys.modules

def _colab_upload_prompt():
    try:
        from google.colab import files
    except Exception:
        raise RuntimeError("google.colab.files not available")
    print("Upload IMDB Dataset CSV now.")
    uploaded = files.upload()
    if not uploaded:
        raise FileNotFoundError("No file uploaded.")
    return list(uploaded.keys())[0]

def _normalize_path(p: str) -> str:
    return None if p is None else p.replace('\\','/')

def safe_read_csv(path: str) -> pd.DataFrame:
    path = _normalize_path(path)
    p = Path(path)
    if not p.exists():
        if _is_colab():
            fname = _colab_upload_prompt()
            path = os.path.abspath(fname)
            p = Path(path)
        else:
            raise FileNotFoundError(f"CSV not found at {path}")
    try:
        df = pd.read_csv(path)
    except Exception:
        # fallback robust attempts
        try:
            df = pd.read_csv(path, engine='python', on_bad_lines='warn')
        except Exception:
            # last resort: try to fix unbalanced quotes
            cleaned = str(Path(path).with_name(Path(path).stem + "_cleaned" + Path(path).suffix))
            with open(path, 'r', errors='replace') as fin, open(cleaned, 'w', errors='replace') as fout:
                buf = ''
                for ln in fin:
                    buf += ln
                    if buf.count('"') % 2 == 0:
                        fout.write(buf); buf = ''
                if buf:
                    fout.write(buf)
            df = pd.read_csv(cleaned, engine='python', on_bad_lines='skip')
    df.columns = [c.strip() for c in df.columns]
    # normalize column names
    cols_l = [c.lower() for c in df.columns]
    if 'review' not in cols_l or 'sentiment' not in cols_l:
        raise ValueError("CSV must contain 'review' and 'sentiment' columns")
    # rename if needed
    rename_map = {}
    for c in df.columns:
        if c.lower()=='review': rename_map[c]='review'
        if c.lower()=='sentiment': rename_map[c]='sentiment'
    df = df.rename(columns=rename_map)
    df = df[['review','sentiment']].copy()
    df['label'] = df['sentiment'].map({'positive':1,'negative':0})
    return df

def train_val_test_split(df: pd.DataFrame, seed=SEED):
    train, test = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=seed)
    train, val = train_test_split(train, test_size=0.1111, stratify=train['label'], random_state=seed)
    return train.reset_index(drop=True), val.reset_index(drop=True), test.reset_index(drop=True)

# -------------------------
# TF-IDF baseline
# -------------------------
def run_tfidf_baseline(train_texts, test_texts, train_labels, test_labels):
    t0 = time.time()
    vec = TfidfVectorizer(max_features=30000, ngram_range=(1,2), stop_words='english')
    X_train = vec.fit_transform(train_texts)
    X_test = vec.transform(test_texts)
    clf = LogisticRegression(max_iter=2000)
    clf.fit(X_train, train_labels)
    train_time = time.time() - t0
    preds = clf.predict(X_test)
    metrics = {
        'accuracy': accuracy_score(test_labels, preds),
        'precision': precision_score(test_labels, preds, zero_division=0),
        'recall': recall_score(test_labels, preds, zero_division=0),
        'f1': f1_score(test_labels, preds, zero_division=0)
    }
    cm = confusion_matrix(test_labels, preds)
    return {'vec': vec, 'clf': clf, 'metrics': metrics, 'confusion_matrix': cm, 'train_time_s': train_time}

# -------------------------
# Plot helpers
# -------------------------
def plot_epoch_series(x, y, title, xlabel='epoch', ylabel=None, show=SHOW_PLOTS):
    plt.figure(figsize=(7,3))
    plt.plot(x, y, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    if ylabel: plt.ylabel(ylabel)
    plt.grid(True)
    if show:
        try: plt.show()
        except Exception: pass
    plt.close()

def plot_trainer_logs(logs, show=SHOW_PLOTS, eval_metrics: dict=None):
    if not logs:
        print("No logs to plot.")
        if eval_metrics:
            # fallback
            if 'eval_loss' in eval_metrics or 'loss' in eval_metrics:
                val = eval_metrics.get('eval_loss', eval_metrics.get('loss', None))
                plot_epoch_series([0,1],[val,val],'Eval Loss (single-point)')
            if 'eval_accuracy' in eval_metrics or 'accuracy' in eval_metrics:
                val = eval_metrics.get('eval_accuracy', eval_metrics.get('accuracy', None))
                plt.figure(); plt.bar(['eval_accuracy'], [val]); plt.ylim(0,1);
                if show: plt.show(); plt.close()
        return

    if isinstance(logs, list):
        df = pd.DataFrame(logs)
    elif isinstance(logs, pd.DataFrame):
        df = logs.copy()
    else:
        try:
            df = pd.DataFrame(logs)
        except Exception:
            df = None

    if df is None or df.empty:
        print("No structured logs timeline available.")
        if eval_metrics:
            plot_trainer_logs(None, show=show, eval_metrics=eval_metrics)
        return

    # aggregate by epoch -> get last row per epoch
    if 'epoch' in df.columns:
        df_epoch = df.groupby('epoch', as_index=False).last()
        if 'eval_loss' in df_epoch.columns:
            plot_epoch_series(df_epoch['epoch'].values, df_epoch['eval_loss'].values, 'Eval Loss')
        if 'eval_accuracy' in df_epoch.columns:
            plot_epoch_series(df_epoch['epoch'].values, df_epoch['eval_accuracy'].values, 'Eval Accuracy')
    else:
        # fallback: look for eval_loss rows
        if 'eval_loss' in df.columns and df['eval_loss'].notna().any():
            rows = df[df['eval_loss'].notna()]
            plot_epoch_series(range(len(rows)), rows['eval_loss'].values, 'Eval Loss (index)')
        if 'eval_accuracy' in df.columns and df['eval_accuracy'].notna().any():
            rows = df[df['eval_accuracy'].notna()]
            plot_epoch_series(range(len(rows)), rows['eval_accuracy'].values, 'Eval Accuracy (index)')

# -------------------------
# DistilBERT finetune
# -------------------------
def prepare_hf_dataset(df: pd.DataFrame, tokenizer, max_length=128):
    ds = HFDataset.from_pandas(df[['review','label']])
    def preprocess(batch):
        return tokenizer(batch['review'], truncation=True, padding='max_length', max_length=max_length)
    ds = ds.map(preprocess, batched=True)
    ds.set_format(type='torch', columns=['input_ids','attention_mask','label'])
    return ds

def _make_training_args_compat(**kwargs):
    sig = inspect.signature(TrainingArguments.__init__)
    params = set(sig.parameters.keys()); params.discard('self')
    filtered = {}
    for k,v in kwargs.items():
        if k in params:
            filtered[k]=v
    # ensure output_dir exists logically (we're not writing, but Trainer validates)
    if 'output_dir' not in filtered:
        filtered['output_dir'] = 'inmemory_output'
    return TrainingArguments(**filtered)

def finetune_distilbert(train_df, val_df, model_name=DISTIL_MODEL, epochs=EPOCHS, batch_size=BATCH_SIZE, lr=LR, max_length=MAX_LENGTH):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
    train_ds = prepare_hf_dataset(train_df, tokenizer, max_length=max_length)
    val_ds = prepare_hf_dataset(val_df, tokenizer, max_length=max_length)

    accuracy = evaluate.load('accuracy')
    precision = evaluate.load('precision')
    recall = evaluate.load('recall')
    f1 = evaluate.load('f1')

    def compute_metrics(pred):
        labels = pred.label_ids
        preds = np.argmax(pred.predictions, axis=1)
        return {
            'eval_accuracy': accuracy.compute(predictions=preds, references=labels)['accuracy'],
            'precision': precision.compute(predictions=preds, references=labels, average='binary')['precision'],
            'recall': recall.compute(predictions=preds, references=labels, average='binary')['recall'],
            'f1': f1.compute(predictions=preds, references=labels, average='binary')['f1']
        }

    training_args = _make_training_args_compat(
        output_dir='inmemory_out',
        num_train_epochs=1,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=min(batch_size*2, 64),
        learning_rate=lr,
        weight_decay=0.01,
        logging_dir='inmemory_logs',
        fp16=torch.cuda.is_available(),
        evaluation_strategy='epoch',   # ask HF to evaluate each epoch (if supported)
        logging_strategy='steps',
        logging_steps=5000,
        save_strategy='no',
        load_best_model_at_end=False
    )

    trainer = Trainer(model=model, args=training_args, train_dataset=train_ds, eval_dataset=val_ds, compute_metrics=compute_metrics)

    history = []
    total_t0 = time.time()
    last_eval_metrics = None

    for ep in range(1, epochs+1):
        print(f"Epoch {ep}/{epochs} - training ...")
        t0 = time.time()
        trainer.train(resume_from_checkpoint=None)
        epoch_time = time.time() - t0

        # get latest train loss if available
        train_loss = None
        logs = getattr(trainer.state, 'log_history', None)
        if isinstance(logs, list) and len(logs)>0:
            for entry in reversed(logs):
                if 'loss' in entry:
                    train_loss = float(entry['loss'])
                    break

        # evaluate
        try:
            eval_metrics = trainer.evaluate(eval_dataset=val_ds)
        except Exception as e:
            print("trainer.evaluate failed:", e)
            eval_metrics = None

        row = {'epoch': ep, 'epoch_time_s': epoch_time, 'train_loss': train_loss}
        if eval_metrics:
            # normalize keys
            if 'eval_loss' in eval_metrics:
                row['eval_loss']=float(eval_metrics['eval_loss'])
            elif 'loss' in eval_metrics:
                row['eval_loss']=float(eval_metrics['loss'])
            if 'eval_accuracy' in eval_metrics:
                row['eval_accuracy']=float(eval_metrics['eval_accuracy'])
            elif 'accuracy' in eval_metrics:
                row['eval_accuracy']=float(eval_metrics['accuracy'])
            last_eval_metrics = eval_metrics

        history.append(row)
        print(f"Finished epoch {ep}: train_loss={train_loss}, eval_metrics_keys={list((eval_metrics or {}).keys())}")

    total_train_time = time.time() - total_t0
    logs_df = pd.DataFrame(history) if len(history)>0 else None

    # plot
    try:
        plot_trainer_logs(logs_df.to_dict(orient='records') if logs_df is not None else None, eval_metrics=last_eval_metrics)
    except Exception as e:
        print("plot_trainer_logs error:", e)

    return {'trainer': trainer, 'tokenizer': tokenizer, 'train_time_s': total_train_time, 'logs_df': logs_df, 'eval_metrics': last_eval_metrics}

# -------------------------
# Embeddings (base) + LR
# -------------------------
def extract_embeddings(model_name, texts: List[str], batch_size=32, max_length=128, device=DEVICE):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name).to(device)
    model.eval()
    embs = []
    with torch.no_grad():
        for i in range(0, len(texts), batch_size):
            chunk = texts[i:i+batch_size]
            toks = tokenizer(chunk, truncation=True, padding='max_length', max_length=max_length, return_tensors='pt')
            toks = {k: v.to(device) for k,v in toks.items()}
            out = model(**toks)
            pooled = out.last_hidden_state[:,0,:].cpu().numpy()
            embs.append(pooled)
    return np.vstack(embs) if len(embs)>0 else np.zeros((0, model.config.hidden_size))

# -------------------------
# GPT-2 prompt scoring
# -------------------------
def gpt2_score_texts(model_name, texts: List[str], device=DEVICE):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
        model.config.pad_token_id = model.config.eos_token_id
    model.eval()
    scores = []
    def seq_logprob(prompt_ids, suffix_ids):
        full_ids = prompt_ids + suffix_ids
        max_len = getattr(tokenizer, "model_max_length", None) or 1024
        if len(full_ids) > max_len:
            needed = max_len - len(suffix_ids)
            prompt_trim = prompt_ids[-needed:] if needed>0 else []
            full_ids = prompt_trim + suffix_ids
        input_ids = torch.tensor([full_ids], dtype=torch.long).to(device)
        with torch.no_grad():
            out = model(input_ids)
            logits = out.logits
            logp = torch.log_softmax(logits, dim=-1)
            targets = input_ids[:,1:]
            selected = logp[:, :-1, :].gather(-1, targets.unsqueeze(-1)).squeeze(-1)
            # compute suffix part only
            suffix_start = len(full_ids) - len(suffix_ids)
            start_idx = max(0, suffix_start - 1)
            relevant = selected[:, start_idx:]
            return float(relevant.sum().item())
    for text in texts:
        pids = tokenizer.encode(text, add_special_tokens=False)
        pos_ids = tokenizer.encode(' positive', add_special_tokens=False)
        neg_ids = tokenizer.encode(' negative', add_special_tokens=False)
        if len(pids) > getattr(tokenizer, 'model_max_length', 1024):
            pids = pids[-getattr(tokenizer, 'model_max_length', 1024):]
        pos_lp = seq_logprob(pids, pos_ids)
        neg_lp = seq_logprob(pids, neg_ids)
        scores.append(pos_lp - neg_lp)
    return scores

# -------------------------
# Small helpers
# -------------------------
def compute_metrics_from_preds(y_true, y_pred):
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall': recall_score(y_true, y_pred, zero_division=0),
        'f1': f1_score(y_true, y_pred, zero_division=0)
    }

def fig_from_confusion(cm, title):
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title(title)
    if SHOW_PLOTS:
        try: plt.show()
        except: pass
    fig = plt.gcf(); plt.close(); return fig

def get_testcases():
    return [
        {'id':'TC1','text':'A wonderful, charming movie  I loved it.','label':1,'notes':'short positive'},
        {'id':'TC2','text':'I wanted to like this film, but it was dull and disappointing.','label':0,'notes':'negation/contrast'},
        {'id':'TC3','text':'If you enjoy painfully slow pacing and wooden acting, this is your masterpiece.','label':0,'notes':'sarcasm/complex'}
    ]

# -------------------------
# Main orchestration (in-memory)
# -------------------------
def run_full_pipeline_inmemory(data_path=DATA_PATH, fast=FAST):
    print('Loading dataset from', data_path)
    df = safe_read_csv(data_path)
    print('Total samples:', len(df))
    if fast:
        df = df.sample(n=2000, random_state=SEED).reset_index(drop=True)
        print('FAST mode: sampled 2000 rows')
    train_df, val_df, test_df = train_val_test_split(df, seed=SEED)
    print('Split sizes:', len(train_df), len(val_df), len(test_df))

    results = {}

    # TF-IDF baseline
    print('Training TF-IDF baseline...')
    tfidf_res = run_tfidf_baseline(train_df['review'].tolist(), test_df['review'].tolist(), train_df['label'].values, test_df['label'].values)
    results['tfidf'] = tfidf_res
    results['tfidf']['confusion_fig'] = fig_from_confusion(tfidf_res['confusion_matrix'], 'TF-IDF Confusion')

    # DistilBERT finetune
    print('Fine-tuning DistilBERT...')
    distil_out = finetune_distilbert(train_df, val_df)
    trainer = distil_out['trainer']; tokenizer = distil_out['tokenizer']
    results['distil_finetuned'] = {'train_time_s': distil_out['train_time_s'], 'trainer': trainer, 'tokenizer': tokenizer}
    # evaluate on test
    print('Evaluating finetuned model on test set...')
    hf_test = HFDataset.from_pandas(test_df[['review','label']])
    def tok_map(batch): return tokenizer(batch['review'], truncation=True, padding='max_length', max_length=MAX_LENGTH)
    hf_test = hf_test.map(tok_map, batched=True)
    hf_test.set_format(type='torch', columns=['input_ids','attention_mask','label'])
    try:
        pred_res = trainer.predict(hf_test)
    except Exception as e:
        print("trainer.predict failed, falling back to manual batching:", e)
        # manual batching
        model = trainer.model.to(DEVICE); model.eval()
        preds = []
        labels = []
        batch_size = 32
        for i in range(0, len(test_df), batch_size):
            chunk = test_df['review'].tolist()[i:i+batch_size]
            enc = tokenizer(chunk, truncation=True, padding='max_length', max_length=MAX_LENGTH, return_tensors='pt')
            enc = {k: v.to(DEVICE) for k,v in enc.items()}
            with torch.no_grad():
                out = model(**enc)
                batch_preds = out.logits.argmax(dim=-1).cpu().numpy().tolist()
            preds.extend(batch_preds)
            labels.extend(test_df['label'].values[i:i+batch_size].tolist())
        import types
        pred_res = types.SimpleNamespace(predictions=np.array([[0]]) , label_ids=np.array(labels))
        pred_res.predictions = np.zeros((len(preds),2)); pred_res.predictions[np.arange(len(preds)), preds] = 1
    preds = np.argmax(pred_res.predictions, axis=1)
    metrics_fin = compute_metrics_from_preds(test_df['label'].values, preds)
    results['distil_finetuned']['metrics'] = metrics_fin
    cm_fin = confusion_matrix(test_df['label'].values, preds)
    results['distil_finetuned']['confusion_fig'] = fig_from_confusion(cm_fin, 'DistilBERT Finetuned Confusion')
    results['distil_finetuned']['logs_df'] = distil_out.get('logs_df')
    results['distil_finetuned']['eval_metrics'] = distil_out.get('eval_metrics')
    print('Finetuned metrics:', metrics_fin)

    # Base embeddings + LR
    print('Extracting base embeddings...')
    X_train_emb = extract_embeddings(DISTIL_MODEL, train_df['review'].tolist(), batch_size=32, max_length=128)
    X_test_emb = extract_embeddings(DISTIL_MODEL, test_df['review'].tolist(), batch_size=32, max_length=128)
    emb_clf = LogisticRegression(max_iter=2000).fit(X_train_emb, train_df['label'].values)
    emb_preds = emb_clf.predict(X_test_emb)
    emb_metrics = compute_metrics_from_preds(test_df['label'].values, emb_preds)
    results['distil_base_emb'] = {'metrics': emb_metrics, 'train_time_s': None}
    results['distil_base_emb']['confusion_fig'] = fig_from_confusion(confusion_matrix(test_df['label'].values, emb_preds), 'Distil Base Emb + LR')
    print('Base emb metrics:', emb_metrics)

    # GPT-2 prompt scoring
    print('Running GPT-2 prompt scoring (sample)...')
    sample_texts = test_df['review'].tolist() if len(test_df)<=500 else test_df['review'].tolist()[:500]
    sample_labels = test_df['label'].values[:len(sample_texts)]
    gpt_scores = gpt2_score_texts(GPT2_MODEL, sample_texts)
    gpt_preds = [1 if s>0 else 0 for s in gpt_scores]
    gpt_metrics = compute_metrics_from_preds(sample_labels, gpt_preds)
    results['gpt_prompt'] = {'metrics': gpt_metrics, 'score_time_s': None}
    results['gpt_prompt']['confusion_fig'] = fig_from_confusion(confusion_matrix(sample_labels, gpt_preds), 'GPT-2 Prompt Confusion')
    print('GPT metrics (sample):', gpt_metrics)

    # Testcases
    tcs = get_testcases(); rows=[]
    for tc in tcs:
        r = {'id':tc['id'],'text':tc['text'],'true_label':tc['label']}
        # tfidf
        r['tfidf_pred'] = int(results['tfidf']['clf'].predict(results['tfidf']['vec'].transform([tc['text']]))[0])
        # finetuned model
        enc = tokenizer([tc['text']], truncation=True, padding='max_length', max_length=MAX_LENGTH, return_tensors='pt')
        enc = {k: v.to(DEVICE) for k,v in enc.items()}
        model = trainer.model.to(DEVICE); model.eval()
        with torch.no_grad():
            out = model(**enc)
            r['distil_finetuned_pred'] = int(out.logits.argmax(dim=-1).cpu().numpy()[0])
        # base emb
        emb = extract_embeddings(DISTIL_MODEL, [tc['text']], batch_size=1, max_length=128)
        r['distil_base_emb_pred'] = int(emb_clf.predict(emb)[0])
        # gpt
        score = gpt2_score_texts(GPT2_MODEL, [tc['text']])[0]
        r['gpt_pred'] = 1 if score>0 else 0
        rows.append(r)
    results['testcases_df'] = pd.DataFrame(rows)
    print(results['testcases_df'])

    # comparative df
    comp = []
    comp.append({'Model':'TF-IDF + LR', **results['tfidf']['metrics']})
    comp.append({'Model':'DistilBERT (finetuned)', **results['distil_finetuned']['metrics']})
    comp.append({'Model':'DistilBERT (base emb) + LR', **results['distil_base_emb']['metrics']})
    comp.append({'Model':'GPT-2 (prompt)', **results['gpt_prompt']['metrics']})
    results['comparative_df'] = pd.DataFrame(comp)
    print('Comparative:')
    print(results['comparative_df'])
    return results

# Run (if executed as script in notebook)
if __name__ == '__main__':
    results = run_full_pipeline_inmemory()
```
